```python
from selenium import webdriver
import pandas as pd
import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.olx.com.ec/autos_c378')
```

```python
time.sleep(3)
autos = driver.find_elements_by_xpath('.//li[@data-aut-id="itemBox"]')
```

```python
time.sleep(3)
nombres = []
precios = []
detalles = []
for auto in autos:
    nombre = auto.find_element_by_xpath('.//span[@data-aut-id="itemTitle"]').text
    precio = auto.find_element_by_xpath('.//span[@data-aut-id="itemPrice"]').text
    detalle = auto.find_element_by_xpath('.//span[@data-aut-id="itemDetails"]').text
    nombres.append(nombre)
    precios.append(precio)
    detalles.append(detalle)
    
```

```python
df = pd.DataFrame(list(zip(nombres, precios, detalles)), columns=['Nombre', 'Precio', 'Detalle'])
df
```
